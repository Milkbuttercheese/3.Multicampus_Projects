```python
#데이터 분석에 사용될 모듈
import numpy as np
import pandas as pd

#지도 서비스에 사용될 모듈
import folium

#기타
import re
import json
import requests
```

```python
#기능 1. json파일에서 정보 불러오기

with open('/Users/wooseongkyun/Datascience/프로젝트1/전국_임상심리사_상담소.json','r',encoding='utf-8') as f:
    json_data = json.load(f)


public_or_private=[]
categories=[]
names=[]
coordinates=[]
phone_numbers=[]
website_urls=[]

#임상심리사 상담소의 경우 위도/경도순으로 옴, 좌표값이 없는 경우 여기서 삭제해야 데이터의 순서가 엉키지 않음
for i in range(len(json_data['정신건강관련기관'])):
    if json_data['정신건강관련기관'][i]['좌표']==None:
        json_data['정신건강관련기관'][i]=[]
    else:
        public_or_private.append(json_data['정신건강관련기관'][i]['공공/민간'])
        categories.append(json_data['정신건강관련기관'][i]['기관구분'])
        names.append(json_data['정신건강관련기관'][i]['기관명'])
        coordinates.append(json_data['정신건강관련기관'][i]['좌표'])
        phone_numbers.append(json_data['정신건강관련기관'][i]['전화번호'])
        website_urls.append(json_data['정신건강관련기관'][i]['홈페이지'])

coordinates
```

```python
#임상심리사 상담소의 경우 위도/경도순으로 옴, 좌표값이 없는 
latitude=[]
longitude=[]
for i in range(len(coordinates)):
    latitude.append(coordinates[i][0])
    longitude.append(coordinates[i][1])
    

print(latitude)
print(longitude)
    
    
```

```python
#맵 만들기
mental_service_map = folium.Map(location = [37.55, 126.98], zoom_start = 12, 
                       tiles = 'Stamen Terrain')

# for문 사용
# 데이터프레임에서 가져올 변수 생성 name, lat, lng
# 가져올 데이터 : zip(df의 인덱스, df의 위도, df의 경도)
# 실행 문장
# folium 모듈의 Marker 함수 사용
# 인자([위도, 경도], popup)
# Marker 뒤에 add_to 함수 사용
# 인자(지도 객체)
for name, lat, long in zip(names, latitude,longitude):
    folium.Marker([lat, long], popup = name).add_to(mental_service_map)

# 지도 저장(seoul_colleges.html)
mental_service_map.save('./metal_service.html')
```
