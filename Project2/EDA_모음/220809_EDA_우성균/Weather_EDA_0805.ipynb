#### 데이터 사용 장단점
1. 최근 5년 전체 데이터 사용
- 장점 : 최근 5년간의 비율을 한눈에 볼 수 있어서 비교하기 좋음
- 단점 : 월별로 구분하기엔 데이터가 너무 많아 그래프 나타내기엔 무리

2. 최근 2년 월별 데이터 사용
- 장점 : 전체보다 좀 더 세세하게 볼 수 있어서 월별로 비교하기엔 좋음
- 단점 : 연도별로 비교할 수가 없고 눈이나 안개 같은 경우엔 값이 없는 경우도 있음,여름이나 겨울 같은 경우 비나 눈이 너무 몰려있는 느낌

#### 1. 맑은 날씨

```python
sunny = pd.DataFrame(accident[accident['기상']=='맑음'])
```

```python
#사고 건수
sunny_2017 = sunny['사고_2017'].sum()
sunny_2018 = sunny['사고_2018'].sum()
sunny_2019 = sunny['사고_2019'].sum()
sunny_2020 = sunny['사고_2020'].sum()
sunny_2021 = sunny['사고_2021'].sum()

print(sunny_2017, sunny_2018, sunny_2019, sunny_2020, sunny_2021)
```

```python
#사망자 수
sunny_die_2017 = sunny['사망_2017'].sum()
sunny_die_2018 = sunny['사망_2018'].sum()
sunny_die_2019 = sunny['사망_2019'].sum()
sunny_die_2020 = sunny['사망_2020'].sum()
sunny_die_2021 = sunny['사망_2021'].sum()

print(sunny_die_2017, sunny_die_2018, sunny_die_2019, sunny_die_2020, sunny_die_2021)
```

```python
sunny_2017 = (sunny['사망_2017'].sum() / sunny['사고_2017'].sum())*100
sunny_2018 = (sunny['사망_2018'].sum() / sunny['사고_2018'].sum())*100
sunny_2019 = (sunny['사망_2019'].sum() / sunny['사고_2019'].sum())*100
sunny_2020 = (sunny['사망_2020'].sum() / sunny['사고_2020'].sum())*100
sunny_2021 = (sunny['사망_2021'].sum() / sunny['사고_2021'].sum())*100

print(sunny_2017, sunny_2018, sunny_2019, sunny_2020, sunny_2021)
```

```python
sunny_total = pd.DataFrame({'맑음':[sunny_2017, sunny_2018, sunny_2019, sunny_2020, sunny_2021]})
sunny_total
```
