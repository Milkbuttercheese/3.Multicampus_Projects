```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(dpi=300)
```

```python
df_elapsed_time= pd.read_csv('./발병_후_응급실_도착_소요시간_현황_시도별__20220625164517.csv',encoding='euc-kr')

df_census=pd.read_csv('2015_to_2020_인구총조사.csv',encoding='euc-kr')
#df_census 테이블에서 총인구/ 시도별 항목만 맞게 추출하기
df_census=df_census[df_census['항목']=="총인구[명]"]
cities=['서울특별시','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시','세종특별자치시','경기도','강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도']



df_total_pop= pd.DataFrame()
for city in cities:
    df_total_pop.insert(cities.index(city),city,(df_census[df_census['행정구역별(읍면동)']==city]).squeeze())

df_total_pop.drop(['행정구역별(읍면동)','항목','단위'],axis=0)
```
