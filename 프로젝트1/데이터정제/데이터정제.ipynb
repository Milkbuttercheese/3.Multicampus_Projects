```python
#파일3. 전국_민간_의료기관 json 불러오고 정제하기
with open('/Users/wooseongkyun/Datascience/프로젝트1/데이터정제/전국_민간_의료기관.json','r',encoding='utf-8') as f:
    json_data = json.load(f)


public_or_private4=[]
categories4=[]
names4=[]
address4=[]
latitude4=[]
longitude4=[]
phone_numbers4=[]
website_urls4=[]

for i in range(len(json_data)):
    #정신병원 민간으로 인덱싱해주기
    if json_data[i]['공공/민간'] == '':
        public_or_private4.append('민간')
    else:
        public_or_private4.append(json_data[i]['공공/민간'])
    categories4.append(json_data[i]['기관구분'])
    names4.append(json_data[i]['기관명'])
    address4.append((json_data[i]['주소']))
    latitude4.append(json_data[i]['latitude'])
    longitude4.append(json_data[i]['longitude'])
    phone_numbers4.append(json_data[i]['전화번호'])
    website_urls4.append(json_data[i]['홈페이지'])

print(public_or_private4)
```

```python
#데이터 합쳐서 정제하기
public_or_private= public_or_private1+public_or_private2+public_or_private3+public_or_private4
categories=categories1+categories2+categories3+categories4
name=names1+names2+names3+names4
address=address1+address2+address3+address4
latitude=latitude1+latitude2+latitude3+latitude4
longitude=longitude1+longitude2+longitude3+longitude4
phone_numbers= phone_numbers1+phone_numbers2+phone_numbers3+phone_numbers4
website_urls= website_urls1+website_urls2+website_urls3+website_urls4


#데이터를 데이터프레임에 담기
col_name=['공공/민간','기관구분','기관명','주소','위도','경도','전화번호','홈페이지']
df=pd.DataFrame([public_or_private,categories,name,address,latitude,longitude,phone_numbers,website_urls],index=col_name)
df=df.transpose()
```

```python
#중복된 주소 확인
print(df[df.duplicated(['주소'])])
#중복된 좌표 확인
print(df[df.duplicated(['위도'])& df.duplicated(['경도']) ])
```

```python
#중복된 주소 및 좌표 삭제
df=df.drop_duplicates(['주소'])
df=df.drop_duplicates(['위도'] and['경도'])
```

```python
#위도 경도 nuermic화
df['위도'] = pd.to_numeric(df['위도'], errors='coerce')
df['경도'] = pd.to_numeric(df['경도'], errors='coerce')
```

```python
df.to_json('전국_정신건강관련기관_위치정보.json',orient='table',force_ascii=False,indent=5)
```
